```python
import seaborn as sns
import numpy as np 
%matplotlib inline 
```

```python
data = sns.load_dataset("tips")
```

```python
data
```

```python
sns.histplot(data.total_bill, kde = True);
```

```python
sns.distplot(data.total_bill);
```

```python
sns.jointplot(x = 'total_bill', y = 'tip' , data = data ) #analisa duas varias ao mesmo tempo. #kind=''
```

```python
sns.jointplot(x = 'total_bill', y = 'tip' , data = data , kind="hex")
```

```python
sns.pairplot(data);
```

```python
sns.pairplot(data , hue ="sex");
```

```python
sns.kdeplot(data['total_bill'])
sns.rugplot(data['total_bill'])  
```

# plot categoricos:

```python
sns.barplot(x="sex" , y="total_bill" , data = data ); #estimativa da mÃ©dia. 
```

```python
sns.barplot(x="sex" , y="total_bill" , data = data, estimator = np.std ); #calcula o desvio padrao 
```

```python
sns.countplot(x="sex", data = data );
```

```python
sns.boxplot(x="day", y="total_bill" , data=data );
```

```python
sns.boxplot(x="day", y="total_bill" ,hue= "sex" , data=data );
```

# regressÃ£o 

```python
sns.lmplot(x="total_bill", y="tip", data = data , hue = "sex" , markers = ["o", "v"] );
```

```python
sns.lmplot(x="total_bill", y="tip", data = data , hue = "sex" , markers = ["o", "v"] , col="time" );
```

```python
sns.lmplot(x="total_bill", y="tip", data = data , hue = "sex" , markers = ["o", "v"] , col="time" , row="day"  );
```
